# Build 1 — Grade Calculator

You already wrote the core logic in Day 4 (`calculate_grade`, `print_result`).  
This version turns it into a **complete interactive script**.

**Requirements:**
- Ask the user how many students to enter
- Loop that many times — collect name and score for each student
- Calculate and display each student's letter grade
- At the end, print a class summary:
  - Highest score
  - Lowest score
  - Class average
  - How many students passed (score >= 60)

**Hint:** Store each student as a dict `{"name": ..., "score": ..., "grade": ...}` in a list.

````python
# Build 1 — Grade Calculator


students = []

def calculate_grade(score):
    if score >= 90:
        return 'A'
    elif score >= 80:
        return 'B'
    elif score >= 70:
        return 'C'
    elif score >= 60:
        return 'D'
    else:
        return 'F'

num_students = int(input("How many students? "))
for i in range(num_students):
    name = input("Student name : ")
    score = float(input("Score 90-100):  "))
    grade = calculate_grade(score)
    students.append({"name": name, "score": score, "grade": grade})

for student in students:
      print(f"{student['name']}: {student['score']} — {student['grade']}")


top = max(students, key=lambda s: s["score"])
print(f"Top student: {top['name']} with {top['score']}")
bottom = min(students, key=lambda s: s["score"])
print(f"Lowest score: {bottom['name']} with {bottom['score']}")
average = sum(s["score"] for s in students) / len(students)
print(f"Class average: {round(average, 1)}")
passed = len([s for s in students if s["score"] >= 60])
print(f"Students who passed: {passed}/{num_students}")
    
    
    
    


````

# Build 2 — Number Guessing Game

**Requirements:**
- Computer picks a random number between 1 and 100
- Player has a limited number of attempts (you decide how many — 7 is fair)
- Each guess: tell the player if the answer is higher or lower
- If they guess correctly: congratulate them and show how many attempts it took
- If they run out of attempts: reveal the number
- Ask if they want to play again

**New concept — the `random` module:**

````python
# Run this cell to see how random works
import random

number = random.randint(1, 100)   # random integer between 1 and 100 (inclusive)
print(f"Random number: {number}")
print(f"Another one: {random.randint(1, 100)}")
````

````python
# Build 2 — Number Guessing Game
import random

attempts = 0
while True:
    answer = input("Do you want to play?(y/n): ")  
    if answer == "y":
        number = random.randint(1,100)
        attempts = 0  # reset here
        while attempts < 7:
            guess = int(input("Your guess: "))
            attempts += 1
            # check guess here
            if guess == number:
               print(f"Correct! You got it in {attempts} attempts.")
               break
            elif guess < number:
                  print("Higher")
            else:
                  print("Lower")
        else:
            # # this runs only if the while loop completed without a break
            print(f"Out of attempts! The number was {number}")
            
    else:
        print("Thanks for playing")
        break
                    


````

# Build 3 — Contact List Manager

A simple command-line app to manage contacts. Think mini address book.

**Requirements:**
- Store contacts as a list of dicts — each contact has: `name`, `phone`, `email`
- Menu loop with these options:
  ```
  1. Add contact
  2. View all contacts
  3. Search by name
  4. Delete contact
  5. Quit
  ```
- Each option calls its own function
- Search should be case-insensitive (hint: use `.lower()`)
- Delete should confirm before removing

**Hint:** Start with the menu loop and stub functions, then fill each one in.

````python
# Build 3 — Contact List Manager
contact= [
    {"name": "Duane", "phone": "6462222222", "email": "db@ai.com"},
    {"name": "steve", "phone": "6462222223", "email": "st@ai.com"},
    {"name": "tony", "phone": "6462222224", "email": "to@ai.com"},  
    {"name": "bing", "phone": "6462222225", "email": "bi@ai.com"},
]

#function to add contact
def add_contact(contact, name, phone, email):
     contact.append({
        "name": name,
        "phone": phone,
        "email": email
    })
#Function to view all contacts
def view_contacts(contact):
    print(f"{'Name':<10} {'Phone':<15} {'Email':<10}")
    print("-" * 35)
    for contacts in contact:
        print(f"{contacts['name']:<10} {contacts['phone']:<15} {contacts['email']:<10}")

#Function to search for particular contacts
def search_contacts(contact, name):
     for contacts in contact:
        if contacts["name"].lower() == name.lower():
            print(f"Name:  {contacts['name']}")
            print(f"Phone: {contacts['phone']}")
            print(f"Email: {contacts['email']}")
            return
     print(f"{name} not found")

def delete_contact(contact, name):
    for contacts in contact:
        if contacts["name"].lower == name.lower():
            contact.remove(contacts)
            print(f"{name} deleted")
            return
    print(f"{name} not found")
    

#Loops through the option, calls function depending on the option
while True:
      print("\n1. Add  2. View  3. Search  4. Delete  5. Quit")
      choice = input("Choose: ")

      if choice == "1":
         name = input("Please enter name: ")
         phone = input("Please enter the phone: ")
         email = input("Please enter the email: ")
         add_contact(contact,name,phone,email)
      elif choice == "2":
          view_contacts(contact)
      elif choice == "3":
          name = input("Enter the name to search: ")
          search_contacts(contact, name)
      elif choice == "4":
          name = input("Enter the name of the contact you want to remove: ")
          answer = input("Are you sure you want to remove the contact: (y/n)")
          if answer == "y":
              delete_contact(contact, name)
          else:
              print("Contact will not be removed")
      elif choice == "5":
          print("Goodbye")
          break
````
